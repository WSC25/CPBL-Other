```python
agg_dict = {
    "打席": "sum",
    "打數": "sum",
    "得分": "sum",
    "安打": "sum",
    "打點": "sum",
    "二安": "sum",
    "三安": "sum",
    "全壘打": "sum",
    "四壞": "sum",
    "觸身": "sum",
    "三振": "sum",
    "盜壘": "sum",
    "盜壘失敗": "sum"
}
batter_total = merged_stat.groupby("名稱").agg(agg_dict).reset_index()
batter_total
```

```python
batter_total['一安'] = batter_total['安打']-batter_total['二安']-batter_total['三安']-batter_total['全壘打']
batter_total['TB'] = batter_total['一安']+batter_total['二安']*2+batter_total['三安']*3+batter_total['全壘打']*4
```

## 代跑次數處理

```python
excel_file_path = 'C:/Users/ASUS/Documents/實習/Rebas/CPBL_2023.xlsx'  # 請替換成你的 Excel 檔案路徑
sheet_name3 = 'AWAY-換人事件'  # 請替換成你要載入的工作表名稱
sheet_name4 = 'HOME-換人事件'
```

```python
away_PR = pd.read_excel(excel_file_path, sheet_name=sheet_name3)
home_PR = pd.read_excel(excel_file_path, sheet_name=sheet_name4)
```

```python
merged_PR = pd.concat([away_PR, home_PR], ignore_index=True)
merged_PR
```

```python
def calculate_so_go_count(series):
    pr_count = (series == 'PR').sum()
    return pd.Series({'PR': pr_count})
```

```python
merged_PR=merged_PR.groupby(['上場選手'])['守位'].apply(calculate_so_go_count).reset_index()
merged_PR = merged_PR.rename(columns={'上場選手':'名稱','level_1': 'PR', '守位': 'PR次數'})
merged_PR
```

```python
merged_PR = merged_PR.drop(['PR'], axis = 1)
merged_PR
```

```python
df = pd.merge(batter_total, merged_PR, on=['名稱'])
```

```python
df['RS%']=100*(df['得分']-df['全壘打'])/(df['安打']+df['觸身']+df['四壞']-df['全壘打']+df['PR次數'])
df['PwrSpd']=(2*df['全壘打']*df['盜壘'])/(df['全壘打']+df['盜壘'])
df['BA'] = df['安打']/df['打數']
df['SecA']=(df['四壞']-df['盜壘失敗']-df['安打']+df['盜壘']+df['TB'])/df['打數']
df['EqA'] = (df['安打']+df['TB']+1.5*(df['四壞']+df['觸身'])+df['盜壘'])/(df['打數']+df['四壞']+df['觸身']+df['盜壘失敗']+df['盜壘']/3)
```
